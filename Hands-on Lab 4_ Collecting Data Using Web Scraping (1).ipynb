## Extract information from the given web site

```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
from bs4 import BeautifulSoup 
import requests
import pandas as pd
!pip install html5lib
```

Download the webpage at the url


```python
#your code goes here
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
data  = requests.get(url).text
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data,"html.parser")
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
table = soup.find('table') # in html table is represented by the tag <table>

#Get all rows from the table
for row in table.find_all('tr'): # in html table row is represented by the tag <tr>
    # Get all columns in each row.
    cols = row.find_all('td') # in html a column is represented by the tag <td>
    color_name = cols[1].getText() # store the value in column 3 as color_name
    color_code = cols[3].getText() # store the value in column 4 as color_code
    print("{}--->{}".format(color_name,color_code))
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here

table_rows = table.find_all('tr')
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
df=pd.DataFrame(l, columns=["column1", "Language", "c3", "Average annual salary", "c5"])
df=df.drop(columns=["column1", "c3", "c5"], axis=1)
df
print(df)

df.to_csv("popular-languages.csv", index=False, sep=';')

import base64,io
from IPython.display import HTML

def create_download_link( df, title = "Download CSV file", filename = "popular-languages"): 
     csv = df.to_csv()
     b64 = base64.b64encode(csv.encode())
     payload = b64.decode()
     html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
     html = html.format(payload=payload,title=title,filename=filename)
     return HTML(html)

create_download_link(df) 
```
